#### **(A) What is the biggest predictor of a large CO2 output per capita of a country?**

To answer this question we must first decide on what predictors we want to research and a time frame. Otherwise it would be far too broad to answer. This analysis includes the following predictors: gross domestic product (GDP) per capita, agricultural land per capita (ha), daily meat consumption per capita (grams) and electricity generation per capita (kWh) over the course of 2000 - 2022. To quantify in how far a predictor predicts CO2 output i the calculated pearson's correlation coefficients between the predictors and the CO2 output. In figure 1 the coefficients for the 4 predictors are shown. Among the predictors GDP per capita seems to be the biggest indication of a large CO2 output per capita with a correlation coefficient of 0.76. The second and third biggest predictors are electricity generation- and daily meat consumption per capita with correlation coefficients of 0.63 and 0.6 respectively. Lastly the predictor agricultural land per capita had a correlation coefficent of 0.1. It is important to note that This analysis does not tell us about correlations between predictors. It is likely however that GDP is correlated with electricity generation as a higher electricity generation would logically indicate more economic activity. In contrast to GDP, agricultural land per capita does not seem to predict a large CO2 output per capita at all. A correlation coefficient of 0.1 is negligible. According to the data many countries actually see a reduction in agriculatural land per capita over the years. This could be due to the fact that the amount of agricultural land stays the same, goes down or goes up very slowly compared to the growth of the population. In conclusion, among the analyzed predictors GDP per capita appears to be the biggest predictor of CO2 output per capita.

#### **(B) Which countries are making the biggest strides in decreasing CO2 output?**

To answer this question we must first decide what time frame we want to look at. I decided to start at 2015 because that is when the paris agreement was signed and ostensibly the year in which most countries started focussing on CO2 reduction. The most recent data encompasses the year 2022 so the timeframe used for this analysis is 2015 - 2022. First I calculated the total CO2 emissions in the years 2015 and 2022 by multiplying the CO2 emission per capita by the population. Next the relative change in total CO2 output from 2015 to 2022 was calculated for each country using the percentage change formula. Lastly I made a bar chart to visualize what countries had the biggest relative decrease in CO2 output. The bar chart is visible in figure 2. As can be seen in the figure Curacao had the largest relative decrease in CO2 output followed by Venezuela. Among the 20 countries with the largest decreases in CO2 output are many European countries as well, such as Estonia, Portugal, Spain, the Netherlands, Germany and the United Kingdom. Interestingly China, Russia and the United States appear to be doing relatively little to combat CO2 emissions. It is important to note that this analysis does not take the reason for the reduction in emissions into account. This means that countries that appear to be making big strides such as Curacao and Venezuela could actually be seeing very large reductions in CO2 emissions because of various reasons. It is, for example, known that Venezuela is suffering a big economic crisis which would influence the industrial activiy in the country and consequently the CO2 emissions. This makes it difficult to say whether Venezuela is making big strides in decreasing CO2 output or not. In general, I feel like it is safe to say that the European countries visible in figure 2 are making a concentrated effort to reduce emissions significantly and are therefore making the biggest strides.

First I plotted the data for the energy sources in a graph, this graph is visible in figure 3. According to the data onshore wind is the cheapest non-fossil fuel energy source as of 2022, closely followed by solar photovoltaic power. Concentrated solar power was the most expensive energy source. The decrease in cost for the energy sources year over year seems to slow down around the years 2018 - 2020.

```python
# What is the biggest predictor of a large CO2 output per capita of a country?
# Find datasets with various possible factors first (on a per capita basis).
# Possible predictors are for example: GDP, meat consumption, agricultural land use and electricity generation.

# Drop empty rows, useless columns and filter data. In this case i decided to drop all rows with a year less than 2000 to save RAM (it crashes otherwise)
dfs = [co2_emissions, gdp, agri, meat, elec]
filtered_dfs = []
for df in dfs:
    df.drop(columns={'Code'}, inplace=True)
    df.dropna(how='all', inplace=True)
    filtered_df = df[df['Year'] >= 2000]
    filtered_dfs.append(filtered_df)

# Merge all the relevant dataframes into one
emission_predictors = filtered_dfs[0]
for df in filtered_dfs[1:]:
    emission_predictors = pd.merge(emission_predictors, df, on=['Entity', 'Year'])

# Calculate the correlation between the predictors and co2 emissions
predictors = ['GDP per capita', 'Agricultural land per capita', 'Daily meat consumption per capita', 'Electricity generation per capita']
correlations = []
for predictor in predictors:
    correlation, _ = pearsonr(emission_predictors[predictor], emission_predictors['CO2 emission per capita'])
    correlations.append(correlation)

# Store the correlations in a dataframe
co2_correlations = pd.DataFrame({
    'Predictor': predictors,
    'Correlation with CO2 emissions per capita': correlations
})

# Visualize the correlation between the predictors and co2 emission with a bar chart
colors = ['blue', 'orange', 'green', 'red', 'purple']
plt.figure(figsize=(10, 8))
plt.bar(co2_correlations['Predictor'], co2_correlations['Correlation with CO2 emissions per capita'], color=colors, width = 0.5)
for i, v in enumerate(co2_correlations['Correlation with CO2 emissions per capita']):
    plt.text(i, v, round(v, 2), ha='center', bbox=dict(boxstyle="round", facecolor='whitesmoke', alpha=0.7))
plt.yticks(np.arange(0, 1.05, 0.05))
plt.grid(axis='y')
plt.xlabel('Predictor', weight='bold')
plt.ylabel('Correlation with CO2 emission', weight='bold')
plt.title('Correlation Between Predictors and CO2 Emission per Capita (2000 - 2022)', weight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


```

```python
# Countries that are making the biggest strides in decreasing CO2 output, i will be looking at the difference between 2015 and 2021 because not all countries have a value for 2022
# First i will make a merged dataframe consisting of the co2 emissions and population change dataframes
co2_pop = pd.merge(co2_emissions, pop, on=['Entity', 'Year'])

# Make sure only countries and all countries are in the dataframe
for country in co2_pop['Entity']:
    if country not in countries:
        co2_pop = co2_pop[co2_pop['Entity'] != country]

# Make copies of dataframes for the year 2021 and 2015 and create columns with total co2 emission per country in 2015 and 2021. Calculate the total CO2 emissions instead of per capita as well.
co2_2015 = co2_pop[co2_pop['Year'] == 2015].copy()
co2_2021 = co2_pop[co2_pop['Year'] == 2021].copy()
co2_2015['Total CO2 emissions_2015'] = round(co2_2015['CO2 emission per capita'] * co2_2015['Population'])
co2_2021['Total CO2 emissions_2021'] = round(co2_2021['CO2 emission per capita'] * co2_2021['Population'])

# Merge the two DataFrames on 'Entity'
co2_comparison = pd.merge(co2_2015, co2_2021, on='Entity')

# Calculate the percentage change in CO2 emissions
co2_comparison['Percentage Change'] = ((co2_comparison['Total CO2 emissions_2021'] - co2_comparison['Total CO2 emissions_2015']) / co2_comparison['Total CO2 emissions_2015']) * 100

# Sort the DataFrame by 'Percentage Change' in descending order and select the top 20
top_decreases = co2_comparison.sort_values('Percentage Change').head(20)

# Create a bar chart to show the countries with the highest decreases in CO2 emissions from 2015 - 2021
plt.figure(figsize=(10, 6))
plt.bar(top_decreases['Entity'], top_decreases['Percentage Change'], color='lightgreen')
for i, v in enumerate(top_decreases['Percentage Change']):
    plt.text(i, v, round(v, 1), ha='center', bbox=dict(boxstyle="round", facecolor='whitesmoke', alpha=0.7), size='small')
plt.xlabel('Country', weight='bold')
plt.ylabel('Percentage Change', weight='bold')
plt.title('Top 20 Countries with Largest Decrease in Total CO2 Emissions (2015-2021)', weight='bold')
plt.ylim(-80, 0)
plt.grid(axis='y')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Make dataframe with only total values for the whole world
energy_cost_w = energy_cost[energy_cost['Entity'] == 'World'].sort_values('Year')

# Make an individual dataframe for every energy source in the dataframe
individual_dfs = {}
for column in energy_cost_w.columns:
    if column not in ['Entity', 'Year']:
        individual_dfs[column] = energy_cost_w[['Entity', 'Year', column]].dropna()
for column in individual_dfs:
    column = individual_dfs[column]

# Plot figure for every source of renewable energy
plt.figure(figsize=(10, 6))
for column in individual_dfs:
    if column not in ['Entity', 'Year']:
        x = individual_dfs[column]['Year']
        y = individual_dfs[column][column]
        plt.plot(x, y, label=column)
        plt.scatter(x, y)
plt.xlabel('Year', weight='bold')
plt.ylabel('Cost ($/kWh)', weight ='bold')
plt.title('Energy Cost Over Time', weight='bold')
plt.legend()
plt.grid(True)
plt.xticks(np.arange(1980, 2022, 5))
plt.yticks(np.arange(0, 1.00, 0.05))
plt.tight_layout()
```

```python
# Which non-fossil fuel energy technology will have the best price in the future?
# Only show the cumulative cost worldwide which accounts for all countries
energy_cost_w = energy_cost[energy_cost['Entity'] == 'World'].sort_values('Year')

# Make an individual dataframe for every energy source in the dataframe so we can loop through them for the plot and put them in a single graph
individual_dfs = {}
for column in energy_cost_w.columns:
    if column not in ['Entity', 'Year']:
        individual_dfs[column] = energy_cost_w[['Entity', 'Year', column]].dropna()
for column in individual_dfs:
    globals()[column] = individual_dfs[column]

# Calculating trend lines between with linear regression and plotting them in a graph
plt.figure(figsize=(10, 6))
model = LinearRegression()
for column in individual_dfs:
    if column not in ['Entity', 'Year']:
        df = individual_dfs[column]
        x = df[df['Year'] >= 1984]['Year'].values.reshape(-1, 1)
        y = df[df['Year'] >= 1984][column].values.reshape(-1, 1)
        model.fit(x, y)
        x_pred = np.arange(1984, 2023).reshape(-1, 1)
        y_pred = model.predict(x_pred)
        intercept = model.intercept_[0]
        slope = model.coef_[0][0]
        formula = f'y = {slope:.4f}x + {intercept:.4f}'
        column_f = f'{column}    {formula}'
        plt.plot(x_pred, y_pred, label=column_f)
        plt.scatter(x_pred, y_pred)
        intercept = model.intercept_[0]
        slope = model.coef_[0][0]
plt.xlabel('Year', weight='bold')
plt.ylabel('Cost ($/kWh)', weight ='bold')
plt.title('Energy Cost Trends Over Time', weight='bold')
plt.legend()
plt.grid(True)
plt.yticks(np.arange(0, 1.00, 0.05))
plt.xticks(np.arange(1984, 2023, 1))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```
